```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)
```

```python
# Define the input and output sizes
input_size = 200
output_size = 2

# Create a neural network with four hidden layers
class Net(nn.Module):
    def __init__(self, input_size, output_size):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(input_size, 120)
        self.fc2 = nn.Linear(120, 80)
        self.fc3 = nn.Linear(80, 50)
        self.fc4 = nn.Linear(50, 20)
        self.fc5 = nn.Linear(20, 5)
        self.fc6 = nn.Linear(5, 2)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))
        x = torch.relu(self.fc4(x))
        x = torch.relu(self.fc5(x))
        x = torch.sigmoid(self.fc6(x))
        return x

# Instantiate the network
net = Net(input_size, output_size)

# Define the loss function and optimizer
loss_fn = nn.BCELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=0.001)

# Train the network
for epoch in range(3000):
    for i in range(len(X)):
        # Forward pass
        y_pred = net(X[i])

        # Calculate the loss
        loss = loss_fn(y_pred, y[i])

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

# Test the network
with torch.no_grad():
    for i in range(len(X)):
        y_pred = net(X[i])
        print(f"Input: {X[i]}, Predicted output: {y_pred}, Actual output: {y[i]}")
```
